##  **Understanding Images**

Digital images are made up of very small units called **pixels**. Every pixel stores **color information**, and when millions of pixels combine together, they form a complete image. A colored image usually contains three color channels: **Blue, Green, and Red (BGR)**. In OpenCV, images are read in **BGR format** instead of RGB format. The quality and size of an image depend on the number of pixels present in it. Higher pixels generally mean **better image quality** and more details.

### An image is just a grid of numbers!

| Concept | Explanation |
|---|---|
| **Pixel** | The smallest unit of an image. Each pixel has a color value. |
| **Color image** | Each pixel has **3 numbers**: Blue, Green, Red (BGR in OpenCV). |

### **❌ Why this DOESN'T work in Jupyter Notebooks**

The function `cv2.imshow()` from OpenCV is designed to open images in a separate desktop window outside the browser.

However, in Jupyter Notebook or JupyterLab environments, especially when running on remote systems like a DGX server, the notebook is accessed through a web browser and does not have direct access to your local screen or desktop environment.

Because of this, `cv2.imshow()` may:

* Crash the notebook kernel
* Freeze without displaying anything

This happens because the code is actually running on the remote DGX server, while you are only viewing the notebook interface in your browser. Since the server cannot open graphical pop-up windows on your local machine, `cv2.imshow()` does not function properly in this environment.

Instead, libraries like `matplotlib` are commonly used in notebooks to display images directly inside the browser.

## **⚠️ Why This Image Looks Weird?**

WWhen an image loaded using OpenCV is displayed directly with Matplotlib, the colors may appear incorrect or unusual. This happens because OpenCV and Matplotlib use different color channel formats to store images.

- OpenCV reads images in BGR format
(Blue → Green → Red)
- Matplotlib displays images in RGB format
(Red → Green → Blue)

Since the color order is different, the red and blue channels get swapped when the image is displayed. As a result, the image colors may look strange, with red objects appearing blue or blue objects appearing red.

To fix this issue, the image must be converted from BGR format to RGB format before displaying it with Matplotlib.

```python
cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

`cv2.cvtColor()` A function used to convert an image from one color format to another
`img` The input image
`cv2.COLOR_BGR2RGB` Specifies that the image should be converted from BGR to RGB format

After this conversion, the image colors appear correctly when displayed using Matplotlib.

````python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
plt.title("Flower Image")
plt.axis("off")
plt.show()
````

##  **Convert to Grayscale**

A grayscale image is an image that contains only shades of gray instead of full colors. Unlike a color image, which uses three color channels (Blue, Green, Red), a grayscale image uses only a single channel to represent brightness.

In a grayscale image:

- 0 represents black
- 255 represents white
- Values between 0–255 represent different shades of gray

Because there is only one channel, grayscale images are simpler and require less memory compared to color images.


````python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()
````

## **📏 Resize an Image**

Resizing is an important step in image processing because images often come in different sizes. Many AI and computer vision models require images to have a fixed width and height before processing.


`cv2.resize(img_RGB, (400, 300))`
This function resizes the image stored in `img_RGB`.

* `cv2.resize()` → Function from OpenCV used to change image dimensions
* `img_RGB` → Original image in RGB format
* `(400, 300)` → New image size

Here:

* `400` → Width of the image
* `300` → Height of the image

The resized image is stored in the variable `resized`.


### **Why Resize Images?**

Image resizing is commonly used because:

* Different images may have different dimensions
* Machine Learning and Deep Learning models usually require fixed-size inputs
* Smaller images reduce memory usage
* Processing becomes faster

````python
resized = cv2.resize(img_RGB, (400, 300))

plt.imshow(resized)
plt.show()
````


## **✂️ Crop an Image**

Cropping means cutting out a specific rectangular region of an image.  
In OpenCV/NumPy, images are just arrays, so **cropping = array slicing!**

```
cropped = image[y_start:y_end, x_start:x_end]
```

Think of it like this:

```
image[rows, columns]
image[top:bottom, left:right]
```

> 💡 **Remember:** Rows = Height = Y axis &nbsp;|&nbsp; Columns = Width = X axis

````python
cropped = img_RGB[100:350, 150:600]

plt.imshow(cropped)
plt.axis("off")
plt.show()
````


## **🖊️ Draw on Image**

OpenCV lets you draw lines, rectangles, circles, and more directly on images. This is super useful for:

- Highlighting detected objects
- Annotating images
- Bounding boxes in object detection

### Key drawing functions:


- `cv2.line(img, start_point, end_point, color, thickness)`
- `cv2.rectangle(img, top_left, bottom_right, color, thickness)`
- `cv2.circle(img, center, radius, color, thickness)`

> ⚠️ **Drawing modifies the image in-place!** Always work on a **copy** so you don't destroy the original.

> 💡 Colors are in **BGR** format! E.g., pure red = `(0, 0, 255)`, pure green = `(0, 255, 0)`, pure blue = `(255, 0, 0)`

````python
draw_img = img.copy()
cv2.line(draw_img, (50, 50), (300, 50), (255, 0, 0), 5)

plt.imshow(cv2.cvtColor(draw_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
````

````python
shape_img = img.copy()
cv2.rectangle(shape_img, (100, 100), (300, 300), (0, 255, 0), -1)
cv2.circle(shape_img, (500, 200), 80, (255, 0, 0), 4)

plt.imshow(cv2.cvtColor(shape_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
````

## **🔤 Add Text on Image**

Adding text is very useful for labeling images, showing results, or creating captions.

`cv2.putText(img, text, origin, font, font_scale, color, thickness)`

| Parameter | Meaning |
|---|---|
| `text` | The string to write |
| `origin` | Bottom-left corner of the text (x, y) |
| `font` | Font style (e.g., `cv2.FONT_HERSHEY_SIMPLEX`) |
| `font_scale` | Size of the font (e.g., 1.0 = normal) |
| `color` | BGR color tuple |
| `thickness` | Stroke width of letters |
| `lineType` | Use `cv2.LINE_AA` for smooth/antialiased text |

````python
text_img = img.copy()
cv2.putText(text_img, "Pallavi ji", (200, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)

plt.imshow(cv2.cvtColor(text_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
````

````python
cap = cv2.VideoCapture("football.mp4")
cap
````

````python
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

fps
````

````python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("football_with_text.mp4", fourcc, fps, (width, height))

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Add text to frame
    cv2.putText(
        frame,
        "Hello",
        (100, 100),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2
    )
````

````python
out.write(frame)
````

````python
cap.release()
out.release()

print("Video saved as football_with_text.mp4")
````
